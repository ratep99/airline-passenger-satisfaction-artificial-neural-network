```python
import seaborn as sns
import matplotlib.pyplot as plt

df_corr = data.corr()
df_corr = df_corr.apply(lambda x: round(x, 2))

fig, ax = plt.subplots(figsize=(20, 20))

# Define a custom color map
cmap = sns.diverging_palette(220, 10, as_cmap=True)
mask = np.triu(np.ones_like(df_corr, dtype=bool))

sns.heatmap(data=df_corr, annot=True,mask=mask, ax=ax, cmap=cmap)
plt.show()
```

```python
data.drop("id",axis=1,inplace=True)
```

```python
data = data.dropna()
```

```python
dups = data.duplicated()
print("Broj duplikata", dups.sum())
```

```python
# one-hot encode all categorical variables except for 'Class'
df = pd.get_dummies(data, columns=['Gender', 'Customer Type', 'Type of Travel', 'satisfaction'])
```

```python
from sklearn.preprocessing import LabelEncoder

# create LabelEncoder object
le = LabelEncoder()

# specify label mapping
label_mapping = {'Eco': 0, 'Eco Plus': 1, 'Business': 2}

# fit_transform the Class column with the specified label mapping
df['Flying Class'] = le.fit_transform(df['Class'].map(label_mapping))
```
